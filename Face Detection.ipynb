```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("face.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier("face.xml")
faces = face_cascade.detectMultiScale(img)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),5)
plt.imshow(img)
plt.show()
```

```python
img = cv2.imread("face.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier("face.xml")
eye_cascade = cv2.CascadeClassifier("eye.xml")
faces = face_cascade.detectMultiScale(img)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),5)
    single_face = img[y:y+h,x:x+w]
    eyes = eye_cascade.detectMultiScale(single_face,1.22)
    for ex,ey,ew,eh in eyes:
        cv2.rectangle(single_face,(ex,ey),(ex+ew,ey+eh),(255,0,0),3)
plt.imshow(img)
plt.show()
```

```python
img = cv2.imread("face.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
face_cascade = cv2.CascadeClassifier("face.xml")
smile_cascade = cv2.CascadeClassifier("smile.xml")
faces = face_cascade.detectMultiScale(img)
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),5)
    single_face = img[y:y+h,x:x+w]
    smiles = smile_cascade.detectMultiScale(single_face,1.7,5)
    for sx,sy,sw,sh in smiles:
        cv2.rectangle(single_face,(sx,sy),(sx+sw,sy+sh),(255,0,0),3)
plt.imshow(img)
plt.show()
```

### Face Detection from Live WebCam

```python
import cv2
cam = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier("face.xml")
while True:
    ret,img = cam.read()
    faces = face_cascade.detectMultiScale(img)
    for x,y,w,h in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),5)
    cv2.imshow("Webcam",img)
    if cv2.waitKey(1)==13:
        break
cam.release()
cv2.destroyAllWindows()       
```

## Object Detection By Color

```python
import cv2
import numpy as np
cam = cv2.VideoCapture(0)
while True :
    ret,img = cam.read()
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    ## red color
    low = np.array([161,155,84])
    high = np.array([179,255,255])
    mask_image = cv2.inRange(hsv,low,high)
    final_image = cv2.bitwise_and(img,img,mask = mask_image)
    cv2.imshow("final_Image",final_image)
    cv2.imshow("Mask_Image",mask_image)
    cv2.imshow("HSV",hsv)
    if cv2.waitKey(0): ##Asci Code for Enter Key
        break
cam.release()
cv2.destroyAllWindows()
```
